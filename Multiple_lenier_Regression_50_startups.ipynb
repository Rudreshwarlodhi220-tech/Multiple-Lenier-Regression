
---

### 🔹 Absolute Difference क्या होता है?
मान लो **Actual value (Y)** = 50 और **Predicted value (Ŷ)** = 45  
Error = \( Y - Ŷ = 50 - 45 = 5 \)  

अगर prediction बड़ा हो गया होता, जैसे 55, तो error = \( 50 - 55 = -5 \)  
लेकिन negative error को समझना मुश्किल होता है, इसलिए हम **absolute value** लेते हैं (मतलब sign हटाकर सिर्फ दूरी देखते हैं)।  
तो error = | -5 | = 5  

👉 Absolute difference = Actual और Predicted के बीच की दूरी, चाहे prediction ऊपर हो या नीचे।

---

### 🔹 Units का मतलब
Error हमेशा उसी unit में होगा जिस unit में तुम्हारा target variable है।  
- अगर तुम **house price (₹ में)** predict कर रहे हो → error भी ₹ में होगा।  
- अगर तुम **weight (kg में)** predict कर रहे हो → error भी kg में होगा।  

इसलिए RMSE और MAE को समझना आसान है क्योंकि वो उसी unit में रहते हैं।

---

### 🔹 Square क्यों करते हैं?
1. **Negative error हटाने के लिए**: अगर सिर्फ difference लेते हैं तो कुछ errors positive होंगे, कुछ negative, और average करने पर cancel हो सकते हैं। Square करने से सब positive हो जाते हैं।  
2. **बड़े errors को ज्यादा importance देने के लिए**:  
   - Example: Error = 10 → Square = 100  
   - Error = 2 → Square = 4  
   मतलब बड़े errors बहुत ज्यादा बढ़ जाते हैं, जिससे model को बड़े गलत predictions avoid करने की motivation मिलती है।
   - Example: Useful जब हम चाहते हैं कि बड़े errors को ज्यादा importance मिले। (MSE)

---

### 🔹 तीनों Metrics का Simple Example
मान लो Actual values = [10, 20, 30]  
Predicted values = [12, 18, 33]  

Errors = [2, -2, 3]  

- **MAE** = Average(|2|, |-2|, |3|) = (2+2+3)/3 = 2.33  
- **MSE** = Average(2², (-2)², 3²) = (4+4+9)/3 = 5.67  
- **RMSE** = √5.67 ≈ 2.38  

👉 देखो:  
- MAE और RMSE लगभग बराबर हैं (units में समझ आते हैं)।  
- MSE बड़ा है क्योंकि square करने से बड़े errors और बड़े हो गए।

---

### 🔹 Summary
- **MAE** → Average गलती, simple और समझने में आसान।  
- **MSE** → बड़े errors को ज्यादा penalize करता है।  
- **RMSE** → Error को उसी unit में बताता है, practical interpretation आसान।

#  --- Model Serialization and Deserialization (Using pickel)---
---

````python
import pickle
````

````python
with open('Pridiction.pkl','wb') as file:
    pickle.dump(train_model,file)
````

````python
with open('Pridiction.pkl','rb') as file:
    model = pickle.load(file)
````

 

---

### 🔹 Process का नाम
- **Serialization** → किसी Python object (जैसे ML model) को binary format में बदलकर file में save करना।  
  👉 यहाँ `pickle.dump()` यही काम कर रहा है।  
- **Deserialization** → उस binary file को वापस पढ़कर original Python object में बदलना।  
  👉 यहाँ `pickle.load()` यही काम कर रहा है।  

इसलिए इस पूरे process को **Model Persistence (Saving & Loading a trained model)** भी कहते हैं।  

---

### 🔹 क्यों ज़रूरी है?
- हर बार model को retrain करने की ज़रूरत नहीं।  
- Time और resources बचते हैं।  
- Trained model को दूसरों के साथ share कर सकते हो।  
- Production में deploy करने के लिए model को save करना ज़रूरी होता है।  

---

### 🔹 Simple Analogy
सोचो तुमने एक किताब पढ़कर notes बनाए।  
- **Serialization (dump)** = Notes को notebook में लिखकर सुरक्षित रखना।  
- **Deserialization (load)** = Notebook खोलकर notes फिर से पढ़ना।  

---

👉 तो तुम्हारे code में जो हुआ उसे हम कहते हैं:  
**Pickle का use करके ML model को serialize (save) और deserialize (load) करना।**

     ```python

     # Load model
     with open("ProfitPrediction.pkl","rb") as file:
         model = pickle.load(file)

     # Prediction
     X_test = [[6],[7]]
     Y_predict = model.predict(X_test)
     print("Predictions:", Y_predict)
     ```
   - अब उसे वही trained model मिल जाएगा जो तुमने बनाया था, और वो बिना retrain किए prediction कर सकता है।
